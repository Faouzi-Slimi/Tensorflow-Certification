```python
# To visulize the predictions let's create a function

 
def plot_decition (model,x,y):
  """Plots the decition boundary """
  # Define the axis
  x_min , x_max = x[:,0].min() - 0.1 , x[:,0].max() + 0.1
  y_min , y_max = x[:,1].min()-0.1 , x[:,1].max() + 0.1
  
  xx , yy = np.meshgrid(np.linspace(x_min,x_max,100),np.linspace(y_min,y_max,100))

  # create x value
  x_in = np.c_[xx.ravel(),yy.ravel()]
  # Make predd
  y_pred = model.predict(x_in)
  # che for multi_class
  if len(y_pred[0]) > 1 :
    print(" Mult_class")
    y_pred = np.argmax(y_pred , axis =1).reshape(xx.shape)
  else :
    print( " Binary ")
    y_pred = np.round(y_pred).reshape(xx.shape) 
  #   Plot the decition boundary
  plt.contourf(xx,yy,y_pred,cmap = plt.cm.RdYlBu,alpha=0.7)
  plt.scatter(x[:,0],x[:,1],c=y,s=40,cmap = plt.cm.RdYlBu)
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())
```

```python
# check out the prdictions our model  is making

plot_decition(model=model_3,x=x,y=y)
```

## Finding the best learning rate

1. A kearning rate ** callback ** 
2. Another model
3. A modified loss curves plot.

```python
# set the random seed
tf.random.set_seed(42)


# 1. create the model

model_5 = tf.keras.Sequential([
                               tf.keras.layers.Dense(4,activation = "relu"),
                               tf.keras.layers.Dense(4,activation = "relu"),
                               tf.keras.layers.Dense(1,activation = "sigmoid")
])

# 2. Compile the model

model_5.compile(loss = tf.keras.losses.BinaryCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(lr = 0.01),
                metrics = ["accuracy"])
# create a learning rate callback

callback = tf.keras.callbacks.LearningRateScheduler(lambda epoch : 1e-4 *10**(epoch/20))


# 3. fit the model

history = model_5.fit(x_train , y_train,epochs = 100 , callbacks = [callback])

```

```python
# Check out the history

pd.DataFrame(history.history).plot(figsize = (10,7),xlabel = "epochs")
```

```python
# Plot the lr versus the loss

lrs = 1e-4*(10**(tf.range(100)/20))

plt.figure( figsize = (10,7) )

plt.semilogx(lrs,history.history["loss"])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning Rtae VS loss")
```

## Working with a larger example ( Multi_class classification ) 

```python
from tensorflow.keras.datasets import fashion_mnist

# The data has alrady bee sorted into train and test sets for as

(train_data,train_labels),(test_data,test_labels) = fashion_mnist.load_data()
```

```python
# check the shape of a single example

train_data[0].shape , train_labels[0].shape
```

```python
# Plot a single sample

plt.imshow(train_data[0])
```

```python
train_data.shape
```

```python
# Build a model

# Normalize the data

train_data_norm = train_data / 255.0
test_data_norm = test_data/255.0

# set the random seed

tf.random.set_seed(42)

# 1. create the model

model_1 = tf.keras.Sequential([
                               tf.keras.layers.Flatten(input_shape=(28,28)),
                               tf.keras.layers.Dense(4,activation = "relu"),
                               tf.keras.layers.Dense(4,activation = "relu"),
                               tf.keras.layers.Dense(10,activation = "softmax")


])

# 2. compile the model

model_1.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["accuracy"])

# 3.Fit the model

model_1.fit(train_data_norm,train_labels,epochs=100,validation_data = (test_data_norm,test_labels))
```

## Finding the ideal learning rate


```python
# Set the random seed

tf.random.set_seed(42)

# 1. create the model

model_1 = tf.keras.Sequential([
                               tf.keras.layers.Flatten(input_shape=(28,28)),
                               tf.keras.layers.Dense(4,activation = "relu"),
                               tf.keras.layers.Dense(4,activation = "relu"),
                               tf.keras.layers.Dense(10,activation = "softmax")


])

# 2. compile the model

model_1.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["accuracy"])

# Create the lr collback

collback = tf.keras.callbacks.LearningRateScheduler(lambda epochs : 1e-4 *10**(epochs/20))

# 3.Fit the model

history = model_1.fit(train_data_norm,train_labels,epochs=100,callbacks = [collback],validation_data = (test_data_norm,test_labels))
```

```python
# Plot th learning rate decay curve

lrs = 1e-3 *(10**(tf.range(100)/20))

plt.semilogx(lrs,history.history[ "loss"])
plt.xlabel("LR")
plt.ylabel("Curve")
plt.title("LR Vs CURVE")
```
